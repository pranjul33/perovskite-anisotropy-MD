```python
import matplotlib.pyplot as plt

def parse_data(filename):
    data = {}
    with open(filename, 'r') as file:
        lines = file.readlines()
        timestep = None
        x_vals = []
        rdf_data = {}
        
        for line in lines:
            if line.startswith("#") or line.strip() == "":
                continue
            
            line_parts = line.split()
            
            # Detect timestep header (two columns, e.g. "2000  somevalue")
            if len(line_parts) == 2:
                timestep = int(line_parts[0])
                x_vals = []
                rdf_data = {}
            
            # Detect RDF data lines (more than 2 columns)
            elif len(line_parts) > 2:
                ncols = len(line_parts)
                # First or second column might be the x value depending on your file structure
                # Check which column is 'r'
                x_val = float(line_parts[1])
                x_vals.append(x_val)
                
                # Build rdf_data dictionary dynamically
                for i in range(2, ncols):
                    col_index = i - 1  # Make it start from 1
                    if col_index not in rdf_data:
                        rdf_data[col_index] = []
                    rdf_data[col_index].append(float(line_parts[i]))
                
                data[timestep] = (x_vals, rdf_data.copy())
    return data


def compare_rdf(filenames, c_myRDF, timesteps):
    for filename in filenames:
        data = parse_data(filename)
        file_label = filename.split('/')[-1]
        
        for timestep in timesteps:
            if timestep in data:
                x_vals, rdf_data = data[timestep]
                if c_myRDF in rdf_data:
                    plt.plot(x_vals, rdf_data[c_myRDF], '-o', label=f'{file_label} - Timestep {timestep}')
                else:
                    print(f"Column {c_myRDF} not found in {file_label} at timestep {timestep}.")
            else:
                print(f"Timestep {timestep} not found in {file_label}.")
    
    plt.xlabel('Distance (r)')
    plt.ylabel(f'RDF c_myRDF[{c_myRDF}]')
    plt.legend()
    plt.title(f'RDF Comparison for c_myRDF[{c_myRDF}] at different timesteps across files')
    path = '/home/pranjul/lammps/pran/analysis/plots/rdf_Pb_Ivac.png'
    plt.savefig(path, dpi=900)
    plt.show()


# --- Input section ---
filenames = [
    #"/home/pranjul/lammps/pran/new_with_anisotropy/300K/X/allrdf/stress.rdf",
     "/home/pranjul/lammps/pran/new_with_anisotropy/300K/Z/allrdf/stress.rdf",
    #"/home/pranjul/lammps/pran/new_with_anisotropy/300K/X/777/stress.rdf",
]

# RDF index mapping (example):
# 2 = Pb–Pb
# 4 = I–I
# 6 = Pb–I
# 8 = Pb–N
# 10 = Pb–C
# 12 = I–N
# 14 = I–C
# 16 = C–N
# 18 = H(N)–I
# 20 = H(C)–I

rdf_val = 18   
c_myRDF_to_plot = rdf_val - 1
timesteps_to_compare = [5000,10000,15000]

compare_rdf(filenames, c_myRDF_to_plot, timesteps_to_compare)
```
